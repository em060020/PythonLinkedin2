## group by

```python
#agrupar por día de la semana y calcular el mínimo retraso
df.groupby(by="DayOfWeek")["ArrDelay"].min()
```

```python
#agrupar por día de la semana y calcular la cantidad
df.groupby(by="DayOfWeek")["ArrDelay"].count()
```

```python
#agrupar por día de la semana y mostrar un resumen
df.groupby(by="DayOfWeek")["ArrDelay"].describe()
```

```python
#sacar el promedio por día de dos variables
df.groupby(by="DayOfWeek")["ArrDelay","DepDelay"].mean()
```

```python
#sacar el rango por día
df.groupby(by="DayOfWeek")["ArrDelay","DepDelay"].max()-df.groupby(by="DayOfWeek")["ArrDelay","DepDelay"].min()
```
